# **Lección 2: Preparación de datos de imagen**

En esta lección, descubrirá cómo preparar datos de imágenes para modelar.

Las imágenes se componen de matrices de valores de píxeles.

Los valores de píxeles suelen ser números enteros sin signo en el rango entre 0 y 255. Aunque estos valores de píxeles se pueden presentar directamente a los modelos de redes neuronales en su formato sin procesar, esto puede generar desafíos durante el modelado, como un entrenamiento del modelo más lento de lo esperado.

En cambio, puede resultar muy beneficioso preparar los valores de los píxeles de la imagen antes del modelado, como simplemente escalar los valores de los píxeles al rango 0-1 para centrar e incluso estandarizar los valores.

Esto se denomina normalización y se puede realizar directamente en una imagen cargada. El siguiente ejemplo usa la librería PIL (la biblioteca de manejo de imágenes estándar en Python) para cargar una imagen y normalizar sus valores de píxeles.

# **Lección 3: Convolucional Neural Network**

En esta lección, descubrirá cómo construir una red neuronal convolucional utilizando una capa convolucional, una capa de agrupación y una capa de salida completamente conectada.

# Capas convolucionales (Convolucioal Layers)
Una convolución es la simple aplicación de un filtro a una entrada que da como resultado una activación. La aplicación repetida del mismo filtro a una entrada da como resultado un mapa de activaciones llamado mapa de características, que indica las ubicaciones y la fuerza de una característica detectada en una entrada, como una imagen.

Se puede crear una capa convolucional especificando tanto el número de filtros a aprender como el tamaño fijo de cada filtro, a menudo llamado forma de núcleo.

# Capas de agrupación (Pooling Layers)
La agrupación de capas proporciona un enfoque para reducir el muestreo de mapas de características al resumir la presencia de características en parches del mapa de características.

Max pulling, o agrupación máxima, es una operación de agrupación que calcula el valor máximo o más grande en cada parche de cada mapa de características.

# Capa de clasificador (Classifier Layer)
Una vez extraídas las características, se pueden interpretar y utilizar para hacer una predicción, como clasificar el tipo de objeto en una fotografía.

Esto se puede lograr al aplanar primero los mapas de características bidimensionales y luego agregar una capa de salida completamente conectada. Para un problema de clasificación binaria, la capa de salida tendría un nodo que predeciría un valor entre 0 y 1 para las dos clases.

# Red neuronal convolucional
El siguiente ejemplo crea una red neuronal convolucional que espera imágenes en escala de grises con el tamaño cuadrado de 256 × 256 píxeles, con una capa convolucional con 32 filtros, cada uno con el tamaño de 3 × 3 píxeles, una capa de agrupación máxima y una salida de clasificación binaria capa.

```
# CNN con un solo convolucional, pooling y la capa de salida
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

# Crear nuestro modelo


```
model = Sequential()
# AGREGAR LA CAPA CONVOLUCIONAL
model.add(Conv2D(32, (3,3), input_shape=(256, 256, 1)))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

# Ejemplo usando un modelo pre entrenado  como un clasificador

```
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
```

# CARGAR LA IMAGEN

```
image =load_img('dog.jpg', target_size=(224, 224))
 # convertir la imagen de pixel a numpy array
image = img_to_array(image)
# reshape del dato del modelo
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

```
# prepara la imagen para  modelo vgg16
image = preprocess_input(image)
# cargar el modelo
model = VGG16()
# Probabilidad de predecir nuestra salida
yhat= model.predict(image)
# convertir laprobabilidad a una clase labels
label = decode_predictions(yhat)
label = label[0][0]
print('%s (%.2f%%)' % (label[1], label[2]*100))
```

# **Lección 05: Modelo de clasificación de imagen de entrenamiento**

En esta lección, descubrirá cómo entrenar y evaluar una red neuronal convolucional para la clasificación de imágenes.

El problema de clasificación de la ropa Fashion-MNIST es un nuevo conjunto de datos estándar que se utiliza en visión artificial y aprendizaje profundo.

Es un conjunto de datos compuesto por 60.000 imágenes cuadradas pequeñas en escala de grises de 28 × 28 píxeles de elementos de 10 tipos de ropa, como zapatos, camisetas, vestidos y más.

El siguiente ejemplo carga el conjunto de datos, escala los valores de píxeles, luego ajusta una red neuronal convolucional en el conjunto de datos de entrenamiento y evalúa el rendimiento de la red en el conjunto de datos de prueba.

El ejemplo se ejecutará en solo unos minutos en una CPU moderna; no se requiere GPU.

```
from keras.datasets import fashion_mnist
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import  Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import  Flatten
```

```
(trainX, trainY), (testX, testY) = fashion_mnist.load_data()
trainX =trainX.reshape((trainX.shape[0], 28,28, 1))
testX = testX.reshape((testX.shape[0], 28,28, 1))

trainX, testX = trainX.astype('float32'), testX.astype('float32')
#normalizo
trainX, testX = trainX/255.0, testX/255.0

trainY, testY = to_categorical(trainY), to_categorical(testY)
#DEFINIR MODELO
model = Sequential()
# AGREGAR LA CAPA CONVOLUCIONAL
model.add(Conv2D(32, (3,3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
model.fit(trainX, trainY, epochs=10, batch_size=32, verbose=2)
loss, acc =model.evaluate(testX, testY, verbose=0)
print(loss, acc)
```

# **Lección 06: Image Augmentation**

En esta lección, descubrirá cómo utilizar el aumento de imágenes.

El aumento de datos de imágenes es una técnica que se puede utilizar para expandir artificialmente el tamaño de un conjunto de datos de entrenamiento creando versiones modificadas de imágenes en el conjunto de datos.

El entrenamiento de modelos de redes neuronales de aprendizaje profundo con más datos puede resultar en modelos más preciso, y las técnicas de aumento pueden crear variaciones de las imágenes que pueden mejorar la capacidad de los modelos de ajuste para generalizar lo que han aprendido a nuevas imágenes.

La librería de redes neuronales de aprendizaje profundo o DL de Keras proporciona la capacidad de ajustar modelos mediante el aumento de datos de imágenes a través de la clase ImageDataGenerator .

El siguiente ejemplo cargará la fotografía como un conjunto de datos y usará el aumento de imagen para crear versiones volteadas y rotadas de la imagen que se pueden usar para entrenar un modelo de red neuronal convolucional.

```
# EJEMPLO DE IMAGE AUMENTATION
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```
# cargar la imagen
img = load_img('bird.jpg')
```

```
# convertiendo para matriz de numpy
data = img_to_array(img)
```

```
# Expandir la dimension
samples = expand_dims(data, 0)
```

```
# Creando la imagen aumentada
datagen= ImageDataGenerator(horizontal_flip=True, vertical_flip=True, rotation_range=90)
```

```
it = datagen.flow(samples, batch_size=1)
```

```
for i in range(9):
  pyplot.subplot(330+1 +i)
  batch = it.next()
  image = batch[0]. astype('uint32')
  pyplot.imshow(image)
pyplot.show()
```
